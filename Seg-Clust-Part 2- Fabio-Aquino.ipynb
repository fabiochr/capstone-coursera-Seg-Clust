```python
import geocoder # import geocoder
# initialize your variable to None
lat_lng_coords = None
import pandas as pd
import numpy as np
import requests
import lxml.html as lh

```

```python
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
page = requests.get(url)           #Create a handle, page, to handle the contents of the website
doc = lh.fromstring(page.content)  #Store the contents of the website under doc
tr_elements = doc.xpath('//tr')    #Parse data that are stored between <tr>..</tr> of HTML

#addr = pd.read_html(url)
```

```python
PC = [] #empty list
i = 0
for t in tr_elements[0]: #For each row, store each first element (header) and an empty list
    i+= 1
    name = t.text_content()
    #print ('%d:" %s "'%(i,name))
    PC.append((name ,[]))
```

```python
#Since out first row is the header, data is stored on the second row
for j in range(1,len(tr_elements)):
    #T is our j'th row
    T=tr_elements[j]
    
    #If row is not of size 10, the //tr data is not from our table 
    if len(T)!=3:
        break
    
    #i is the index of our column
    i=0
    
    #Iterate through each element of the row
    for t in T.iterchildren():
        data=t.text_content() 
        #Check if row is empty
        #if i>0:
        #Convert any numerical value to integers
            #try:
                #data=int(data)
            #except:
                #pass
        #Append the data to the empty list of the i'th column
        PC[i][1].append(data)
        #Increment i for the next column
        i+=1
```

```python
Dict={title:column for (title,column) in PC}
df=pd.DataFrame(Dict)
```
